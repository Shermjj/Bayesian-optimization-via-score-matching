```python
from sm_bbo.lfbo import run_lfbo
from sm_bbo.functions import Rosenbrock, Ackley, Rastrigin
import numpy as np
import torch
import pickle
torch.set_default_dtype(torch.float32)
```

```python
param_dims = [5, 10, 20]  # 3 different param_dim values
seeds = range(10)  # 10 different seeds
results = {}

for param_dim in param_dims:
    results[param_dim] = {}
    for seed in seeds:
        np.random.seed(seed)
        torch.manual_seed(seed)

        X_obs = np.random.uniform(-5, 10, size=(4, param_dim))
        X_obs = torch.tensor(X_obs)
        f = Rosenbrock(dim=param_dim, noisy=True)

        X_obs = run_lfbo(X_obs, f, param_dim, n_iter=250)
        fn_evals = f(X_obs)

        running_min = torch.minimum(fn_evals, torch.cummin(fn_evals, dim=0)[0]).numpy()
        results[param_dim][seed] = running_min

# Now 'results' is a nested dictionary that can be easily stored to disk
# It's structured as: {param_dim: {seed: running_min}}

# Save the results dictionary to disk
with open('results_lfbo_rosenbrock.pkl', 'wb') as f:
    pickle.dump(results, f)
```
