```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from matplotlib.ticker import ScalarFormatter
```

```python
with open('results_lfbo_rosenbrock.pkl', 'rb') as f:
    loaded_results_lfbo = pickle.load(f)

with open('results_rs_rosenbrock.pkl', 'rb') as f:
    loaded_results_rs = pickle.load(f)
    
with open('results_sm_bbo_rosenbrock.pkl', 'rb') as f:
    loaded_results_sm = pickle.load(f)
    
plt.rcParams.update({
    "font.size": 10,
    "axes.labelsize": 12,
    "axes.titlesize": 12,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "legend.fontsize": 9,
    "figure.titlesize": 14
})

fig, axes = plt.subplots(1, len(loaded_results_lfbo.keys()), figsize=(12, 4), dpi=300)

methods = {
    'LFBO': loaded_results_lfbo,
    'RS': loaded_results_rs,
    'SM': loaded_results_sm
}

colors = {'LFBO': '#1f77b4', 'RS': '#2ca02c', 'SM': '#d62728'}
linestyles = {'LFBO': '-', 'RS': '--', 'SM': '-.'}

max_iterations = 250

for i, (param_dim, ax) in enumerate(zip(loaded_results_lfbo.keys(), axes)):
    for method, results in methods.items():
        runs = np.array(list(results[param_dim].values()))
        runs = runs[:, :max_iterations]
        mean = np.mean(runs, axis=0)
        std = np.std(runs, axis=0)
        x = np.arange(1, max_iterations + 1)
        
        ax.plot(x, mean, label=method, color=colors[method], linestyle=linestyles[method], linewidth=1.5)
        ax.fill_between(x, mean-std, mean+std, alpha=0.15, color=colors[method])
    
    ax.set_title(f"Dimension = {param_dim}")
    ax.set_xlabel("No. of function evaluations")
    ax.set_ylabel("Immediate Regret (Log)")
    ax.set_yscale('log')
    ax.legend(frameon=False)
    ax.grid(True, linestyle=':', alpha=0.7)
    ax.set_xlim(1, max_iterations)

    y_min, y_max = ax.get_ylim()
    ax.set_ylim(y_min * 0.9, y_max * 1.1)

plt.tight_layout()
fig.subplots_adjust(top=0.88)

# Uncomment the following line to save the figure
plt.savefig('optimization_comparison.pdf', format='pdf', bbox_inches='tight')

#plt.show()
```
